# find all the null values in dataset if any and if found replace the null values with mean of that column

```python
dataset['Cylinders'].fillna(dataset['Cylinders'].mean(),inplace=True)
```

```python
dataset['EngineSize'].fillna(dataset['EngineSize'].mean(),inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset['MSRP']=dataset['MSRP'].str.replace('$','')
dataset['MSRP']=dataset['MSRP'].str.replace(',','')
dataset['MSRP']= pd.to_numeric(dataset['MSRP'])
dataset['MSRP'].mean()
```
